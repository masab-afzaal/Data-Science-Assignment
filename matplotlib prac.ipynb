```python
import matplotlib.pyplot as plt
y1 =[]
y2 =[]
x = range(-100,100,10)
for i in x: y1.append(i**2) 
for i in x: y2.append(-i**2) 

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2000, 2000)
plt.axhline(0) 
plt.axvline(0) 

#plt.savefig("quad.png")

plt.show()
```

```python
x= [1,2,3]
y= [2,4,1]
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("this is simple Graph")
```

```python
import matplotlib.pyplot as plt
x1 = [1,2,3]
y1 = [2,4,1]
plt.plot(x1,y1,label = "plot 1")
x2 = [1,2,3]
y2 = [4,1,3]
plt.plot(x2,y2, label = "plot 2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("this is 2-line Graph")
plt.legend()
plt.show()
```

```python
x= [2,4,6,8]
y= [6,2,8,4]
plt.plot(x,y, color= 'green', linestyle = 'dashed', linewidth = 2, marker = '*', markerfacecolor = 'red', markersize= 12)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("Customize Graph")
plt.show()
```

```python
val = [6,2,7,5,3]
nam = ['A','B','C','D','E']
plt.bar(nam,val, color = 'brown')
plt.show()
```

```python
val = [5,2,8,4,3]
nam = ['A','B','C','D','E']
color = ['green', 'brown']
plt.barh(nam,val, color = color)
plt.show()
```

```python
ages=[2,5,70,40,30,45,43,40,44,60,7,57,18,90,77,32,34,22,26,21,12,29,27,25,21,20,40] 
  
range = (0, 100) 
bins = 10

plt.hist(ages, bins, range, color='purple',histtype='bar',rwidth=0.8) 
  

plt.xlabel('age') 

plt.ylabel('No. of people') 

plt.title('Histogram') 


plt.show()
```

```python
import matplotlib.pyplot as plt
  
activities = ['eat', 'sleep', 'work', 'play'] 
slices = [6, 7, 12, 6]  
colors = ['r', 'y', 'g', 'b'] 

plt.pie(slices, labels = activities, colors=colors,  
         startangle=180, explode = (0, 0, 0.1, 0), 
        radius = 1.2, autopct = '%1.1f%%') 
  
plt.legend() 
plt.show()
```
